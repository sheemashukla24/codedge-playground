# Pandas & DataFrames
Let's explore a small dataset of student scores using pandas.

```python
import pandas as pd

data = {
    'name': ['Aditi', 'Rahul', 'Priya', 'Karan', 'Sneha'],
    'python_score': [88, 72, 95, 60, 84],
    'ml_score': [91, 68, 89, 55, 79]
}

df = pd.DataFrame(data)
df
```

## Basic operations
Try changing the column name or the condition below and re-run the cell.

```python
# Average score
print('Average Python score:', df['python_score'].mean())

# Filter students who scored above 80 in ML
top_students = df[df['ml_score'] > 80]
top_students
```

```python
# Sort by python_score, descending
df.sort_values('python_score', ascending=False)
```
