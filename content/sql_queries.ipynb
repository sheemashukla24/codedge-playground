# SQL with sqlite3
Python's built-in `sqlite3` module lets us run real SQL, right here in the browser.

```python
import sqlite3

conn = sqlite3.connect(':memory:')
cur = conn.cursor()

cur.execute('''
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    course TEXT,
    score INTEGER
)
''')

students = [
    (1, 'Aditi', 'Data Science', 88),
    (2, 'Rahul', 'Machine Learning', 72),
    (3, 'Priya', 'Data Science', 95),
    (4, 'Karan', 'AI', 60),
    (5, 'Sneha', 'Machine Learning', 84),
]

cur.executemany('INSERT INTO students VALUES (?, ?, ?, ?)', students)
conn.commit()
print('Table created and populated.')
```

## Try some queries
Change the query below and re-run to explore.

```python
cur.execute('SELECT name, course, score FROM students WHERE score > 75 ORDER BY score DESC')
for row in cur.fetchall():
    print(row)
```

```python
# Average score per course
cur.execute('SELECT course, AVG(score) FROM students GROUP BY course')
for row in cur.fetchall():
    print(row)
```
